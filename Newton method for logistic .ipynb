First of all, we import numpy library.

```python
import numpy as np
```

Then, we write a function to read the data.
The links of data:
• http://cs229.stanford.edu/ps/ps1/logistic_x.txt
• http://cs229.stanford.edu/ps/ps1/logistic_y.txt

```python
def read_input():
    X_file = open('logistic_x.txt', 'r')
    X = []
    for line in X_file:
        x = [1.] + map(float, line.strip().split())
        X.append(x)
    X_file.close()
    Y_file = open('logistic_y.txt', 'r')
    y = []
    for line in Y_file:
        y_value = float(line.strip())
        y.append(y_value)
    Y_file.close()
    return np.array(X), np.array(y)
```

Now we call read_input() function and store the in 2 numpy array: X and y

```python
X, y = read_input()
```

We set the zero vector as initial theta.

```python
theta = np.array([0. for i in range(len(X[0]))])
theta
```

We write the portion code to compute optimal theta by using Newton method.


<img src="https://i.imgur.com/ECn5vpW.jpg?1">

```python
theta = np.array([0. for i in range(len(X[0]))])
max_iter = 20
for i in range(max_iter):
    margins = y * X.dot(theta)
    #print margins
    probs = 1. / (1 + np.exp(margins))
    grad = -1. / len(X) * (X.T.dot(probs * y))
    H = (1. / len(X)) * (X.T.dot(np.diag(probs * (1 - probs)))).dot(X)
    theta = theta - np.linalg.pinv(H).dot(grad)
```
